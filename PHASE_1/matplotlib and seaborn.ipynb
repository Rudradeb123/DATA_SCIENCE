```python
df["Publisher"].value_counts()[:5]
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x="Publisher",data=df,hue="Platform")
```

```python
top_3_genre=df["Genre"].value_counts().index[:3]
top_3_genre
```

```python
top_3_platforms = df["Platform"].value_counts().index[:3]
top_3_platforms
```

```python
top_3_publisher = df["Publisher"].value_counts().index[:3]
top_3_publisher
```

```python
#combine all
top_3_data=df.loc[
    (df['Publisher'].isin(top_3_publisher)) &
    (df['Platform'].isin(top_3_platforms)) &
    (df['Genre'].isin(top_3_genre))
]
top_3_data
```

```python
#dodge plot
sns.countplot(data=top_3_data,x="Publisher",hue="Platform")
```

```python
#create dataframe to pivot table
df_stacked_plt=pd.crosstab(index=top_3_data["Publisher"],columns=top_3_data["Platform"])
df_stacked_plt
```

```python
#stacked plot
df_stacked_plt.plot(kind="bar",stacked=True)
```

```python
 #create boxplot
sns.boxplot(x="Publisher",y="Global_Sales",data=top_3_data)
```
