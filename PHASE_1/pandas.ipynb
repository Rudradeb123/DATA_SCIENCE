```python
#merging and combining
#customer dataset
df_customers=pd.DataFrame({
    "Customer_ID":[1,2,3],
    "Name":["Rudra","Abhimanyu","Madhab"]
})

#order details
df_orders=pd.DataFrame({
    "Customer_ID":[1,2,4],
    "Order_amount":[205,450,350]
})

print("INNER: ")
print(pd.merge(df_customers,df_orders,on="Customer_ID",how="inner"))

print("OUTER: ")
print()

print(pd.merge(df_customers,df_orders,on="Customer_ID",how="outer"))

print("LEFT: ")
print()

print(pd.merge(df_customers,df_orders,on="Customer_ID",how="left"))

print("RIGHT: ")
print()

print(pd.merge(df_customers,df_orders,on="Customer_ID",how="right"))

print("CROSS:")
print()

print(pd.merge(df_customers, df_orders, how="cross"))
```

```python
# #concatination
# vertically("Row_wise")
# horizontally("column_wise")
# syntax-> pd.concat([df1,df2],axis=0,ignore_index=True)
# ignore_index=True index reset
```

```python
df_customers=pd.DataFrame({
    "Customer_ID":[1,2],
    "Name":["Rudra","Abhimanyu",]
})

#order details
df_orders=pd.DataFrame({
    "Customer_ID":[3,4],
    "Name":["Madhab","Bhim",]
})
print(pd.concat([df_customers,df_orders],axis=0,ignore_index=True))
print()
print(pd.concat([df_customers,df_orders],axis=1,ignore_index=True))
```
